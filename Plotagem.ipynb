```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# dados em colunas
with open("arquivo.txt" , 'r') as file:
    lines = file.readlines()
  
beats = []
time = []
i = 0
for line in lines:
  try:
    i += 1
    beats.append(float(line.strip()))
    time.append(i)
  except:
    pass
```

```python
plt.figure(figsize=(20, 10))
plt.title("Batimento x tempo")
plt.plot(time, beats)
plt.show()
```
